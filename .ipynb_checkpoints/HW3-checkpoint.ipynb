```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
path = "./movies.csv"
df = pd.read_csv(path, header=None)
print("Read {} rows.".format(len(df)))
```

```python
df.columns = ["name","class","date","intro"]
df.head()
```

```python
x = 0
for i in df["class"]:
    df["class"][x] = i.replace("[","").replace("]","").replace("\'","").split(',')[0]
    x = x+1
df
```

```python
import jieba
import jieba.analyse
import math
from collections import Counter
```

```python
jieba.set_dictionary('dict.txt.big')
stopwords = [line.strip() for line in open("stop_words.txt",encoding="utf-8").readlines()]
```

```python
def seg_sentence(sentence):
    sentence_seged = jieba.cut(sentence.strip())
    outstr = ''
    for word in sentence_seged:
        if word not in stopwords:
            if word != '\t':
                outstr += word
                outstr += " "
    return outstr
```

```python
df['intro']=pd.DataFrame(df['intro'].astype(str))
df['intro'] = df['intro'].apply(seg_sentence)
```

```python
df
```

```python
tmp = []
c=0
for i in df['class']:
    if i == '':
        tmp.append(c)
    c = c + 1
df=df.drop(tmp)
```

```python
dd = []
c=0
count=0
for i in df['class']:
    if count == 86:
        break
    if i == '劇情':
        dd.append(c)
        count = count+1
    c = c + 1
dd
df = df.drop(dd)
```

```python
df
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['intro'].values.astype('U'))
print(X.shape)
```

```python
r = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())
print("CountVector")
r
```

```python
from sklearn import preprocessing
nmp=r.to_numpy()
nmp
```

```python
label = df['class'].to_numpy()
label=label.astype('U')
label
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, label,test_size=0.0625,random_state=1)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
accuracy = []
for k in range(1, 100):
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy.append(metrics.accuracy_score(y_test, y_pred))
k_range = range(1,100)
plt.plot(k_range, accuracy)
plt.show()
```

```python
clf=KNeighborsClassifier(n_neighbors=71,p=2,weights='distance',algorithm='brute')
clf.fit(X_train,y_train)
```

```python
clf.score(X_train,y_train)
```

```python
print("準確率:",clf.score(X_test,y_test))
```
